```python
import numpy as np
import pandas as pd
```

```python
top_left = pd.read_csv('top1-25-1.csv')
top_right = pd.read_csv('top1-25-2.csv')
bottom_left = pd.read_csv('top26-50-1.csv')
bottom_right = pd.read_csv('top26-50-2.csv')
```

```python
bottom_part = pd.concat([top_left, top_right], axis=1)
bottom_part = bottom_part.reset_index()
bottom_part.drop(columns='index', inplace=True)
bottom_part = bottom_part.loc[:,~bottom_part.columns.duplicated()]
bottom_part
```

```python
upper_part = pd.concat([bottom_left, bottom_right], axis=1)
upper_part = upper_part.reset_index()
upper_part.drop(columns='index', inplace=True)
upper_part
```

```python
result = pd.concat([upper_part, bottom_part])
result.reset_index(inplace = True)
result.drop(columns='index', inplace=True)
result.index = np.arange(1, len(result)+1)
result
```
